## after reviewing df_1, i begin cleaning it.

```python
df_1.columns
```

```python
cars_accidents = df_1.copy()
cols_1 = ['DATAFLOW',
          'FREQ',
          'OBS_STATUS',
          'NOTE_DS',
          'NOTE_REF_AREA',
          'NOTE_DATA_TYPE',
          'NOTE_RESULT',
          'NOTE_TIME_PERIOD',
          'BASE_PER',
          'UNIT_MEAS',
          'UNIT_MULT']
cars_accidents.drop(columns=cols_1, inplace=True)
cars_accidents.shape
```

```python
cars_accidents.info()
```

```python
cars_accidents.sample(10)
```

```python
cars_accidents = cars_accidents.rename(columns={'REF_AREA': 'Id_code','DATA_TYPE':'Type','RESULT':'Outcome','TIME_PERIOD':'Year','OBS_VALUE':'Obs_value'})
cars_accidents.head(10)
```

```python
cars_accidents['Outcome'] = cars_accidents['Outcome'].replace('9','')
cars_accidents['Metric'] = cars_accidents['Type'] + cars_accidents['Outcome']
cars_accidents.head(10)
```

```python
cars_accidents = cars_accidents.drop(columns=['Type','Outcome'])
cars_accidents['Metric'] = cars_accidents['Metric'].replace({'KILLINJF':'injured', 'KILLINJM' : 'deaths', 'ROADACC' : 'road_acc'})
cars_accidents = cars_accidents[['Id_code','Year','Metric','Obs_value']]
cars_accidents.head(10)
```

```python
cars_accidents_2021_2025 = cars_accidents.pivot_table(index=['Id_code','Year'],columns=['Metric'], values='Obs_value').reset_index()
cars_accidents_2021_2025.columns.name = None
c = ['deaths', 'injured', 'road_acc']
cars_accidents_2021_2025[c] = cars_accidents_2021_2025[c].astype('int64')
cars_accidents_2021_2025.sample(10)
```

```python
comuni_italiani = df_2.copy()
```

```python
comuni_italiani['Superficie (Kmq)'] = comuni_italiani['Superficie (Kmq)'].str.replace('.', '').str.replace(',', '.').astype('float64')
comuni_italiani['Superficie (Kmq)'].dtype
comuni_italiani['Superficie (Kmq)'].sample(10)
```

## I discovered that the single municipality labeled "NaN" (1168) actually corresponds to a town in the province of Turin called None! xD

```python
comuni_italiani['Comune'] = comuni_italiani['Comune'].fillna('None')
```

```python
comuni_italiani[comuni_italiani['Comune'] == 'None']
```

```python
region = {
    1: 'Piemonte',
    2: "Valle d'Aosta",
    3: 'Lombardia',
    4: 'Trentino-Alto Adige',    
    5: 'Veneto',            
    6: 'Friuli-Venezia Giulia',
    7: 'Liguria',
    8: 'Emilia-Romagna',
    9: 'Toscana',
    10: 'Umbria',
    11: 'Marche',
    12: 'Lazio',
    13: 'Abruzzo',
    14: 'Molise',            
    15: 'Campania',
    16: 'Puglia',
    17: 'Basilicata',
    18: 'Calabria',
    19: 'Sicilia',
    20: 'Sardegna'}

comuni_italiani['Regione'] = comuni_italiani['Codice Regione'].map(region)
comuni_italiani.sample(10)
```

```python
rip_geo = {1: 'Nord-Ovest',
           2: 'Nord-Est',
           3: 'Centro',
           4: 'Sud',
           5: 'Isole'}
comuni_italiani['Ripartizione Geografica'] = comuni_italiani['Codice Ripartizione geografica'].map(rip_geo)
comuni_italiani.sample(10)
```

```python
comuni_italiani['Capoluogo di Provincia/Uts'] = comuni_italiani['Capoluogo di Provincia/Uts'].replace({0: 'No', 1: 'Yes'})
comuni_italiani['Capoluogo di Regione'] = comuni_italiani['Capoluogo di Regione'].replace({0: 'No', 1: 'Yes'})
comuni_italiani.sample(10)
```

```python
comuni_italiani.columns
```

```python
comuni_italiani_clean = comuni_italiani[['Codice Ripartizione geografica',
                                         'Codice Regione',
                                         'Codice Comune (alfanumerico)',
                                         'Codice Comune (numerico)',
                                         'Ripartizione Geografica',
                                         'Regione',
                                         'Sigla automobilistica',
                                         'Comune',
                                         'Capoluogo di Provincia/Uts',
                                         'Capoluogo di Regione',
                                         'Popolazione legale',
                                         'Superficie (Kmq)',
                                         ]]

comuni_italiani_clean.sample(10)
```

```python
cars_accidents_2021_2025.to_csv("cars_accidents_2021_2025.csv", index=False)
comuni_italiani_clean.to_csv("comuni_italiani_clean.csv", index=False)
```
