```python
df_1.columns
```

```python
cars_accidents = df_1.copy()
cols_1 = ['DATAFLOW',
          'FREQ',
          'OBS_STATUS',
          'NOTE_DS',
          'NOTE_REF_AREA',
          'NOTE_DATA_TYPE',
          'NOTE_RESULT',
          'NOTE_TIME_PERIOD',
          'BASE_PER',
          'UNIT_MEAS',
          'UNIT_MULT']
cars_accidents.drop(columns=cols_1, inplace=True)
cars_accidents.shape
```

```python
cars_accidents.info()
```

```python
cars_accidents.sample(10)
```

```python
cars_accidents = cars_accidents.rename(columns={'REF_AREA': 'id_code','DATA_TYPE':'type','RESULT':'outcome','TIME_PERIOD':'year','OBS_VALUE':'obs_value'})
cars_accidents.head(10)
```

```python
cars_accidents['outcome'] = cars_accidents['outcome'].replace('9','')
cars_accidents['metric'] = cars_accidents['type'] + cars_accidents['outcome']
cars_accidents.head(10)
```

```python
cars_accidents = cars_accidents.drop(columns=['type','outcome'])
cars_accidents['metric'] = cars_accidents['metric'].replace({'KILLINJF':'injured', 'KILLINJM' : 'deaths', 'ROADACC' : 'road_acc'})
cars_accidents = cars_accidents[['id_code','year','metric','obs_value']]
cars_accidents.head(10)
```

```python
cars_accidents_2021_2025 = cars_accidents.pivot_table(index=['id_code','year'],columns=['metric'], values='obs_value').reset_index()
cars_accidents_2021_2025.columns.name = None
c = ['deaths', 'injured', 'road_acc']
cars_accidents_2021_2025[c] = cars_accidents_2021_2025[c].astype('int64')
cars_accidents_2021_2025.sample(10)
```
